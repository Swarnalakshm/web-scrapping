```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as sa
```

## Scrapping Website

```python
url = "https://www.flipkart.com/search?q=watches+men&sid=r18%2Cf13&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_na&as-pos=1&as-type=RECENT&suggestionId=watches+men%7CWrist+Watches&requestId=71243e4f-8c86-4dc7-85dc-0051949de99f&as-searchtext=watch"
```

## Website response

```python
res = requests.get(url)
```

```python
res
```

## parsing

```python
soup = BeautifulSoup(res.text)
```

```python
soup
```

## watch name

```python
watch = soup.select('._2WkVRV')
```

```python
watch
```

```python
prices = soup.select('div~ div+ div ._373qXS ._30jeq3 , .e30oNt ._30jeq3')
```

```python
prices
```

```python
watch_names=[]
```

```python
for item in watch:
    watch_names.append(item.text)
```

```python
watch_names
```

```python
price_title=[]
```

```python
for item in  prices:
    price_title.append(item.text)
```

```python
price_title
```

## Data Frame

```python
df=sa.DataFrame(watch_names,columns=['watch_names'])
```

```python
df
```

```python
df['price_title']=price_title
```

```python
df
```

```python
df.to_csv('sl.csv')
```
